```python

# ============================================================
# cnn_only.py
# - Strategies: Original, SMOTE, ADASYN
# - Model: CNN fixed architecture (no WOA)
# - IMPORTANT: Oversampling only on TRAIN set (avoid leakage)
# ============================================================

from __future__ import annotations

import os
import random
from pathlib import Path
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import umap

from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

from imblearn.over_sampling import SMOTE, ADASYN

import tensorflow as tf
from tensorflow.keras import layers, models, Model

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    silhouette_score, davies_bouldin_score, calinski_harabasz_score,
)

# ======================
# CONFIG
# ======================
API_KEY = "sMWxUUzNkgFJyEegqotWyVYFJEtK4N7K"  # isi API key MP kamu
TARGET_SYSTEMS = ["cubic", "monoclinic", "orthorhombic", "tetragonal"]

TWO_THETA_MIN = 10
TWO_THETA_MAX = 90
N_POINTS = 2048
MAX_PER_CLASS = 5000

HOME_DIR = Path.home()
CIF_SAVE_DIR = HOME_DIR / "Downloads" / "cif_perovskite"

SEED = 42

EPOCHS_FINAL = 30
BATCH_SIZE = 32
VAL_SPLIT = 0.2


# ======================
# SEED SETTER
# ======================
def set_seed(seed=SEED):
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)


# ======================
# DATA LOADING
# ======================
def load_perovskite_xrd_from_mp(
    api_key: str,
    max_per_class: int = MAX_PER_CLASS,
    two_theta_min: float = TWO_THETA_MIN,
    two_theta_max: float = TWO_THETA_MAX,
    n_points: int = N_POINTS,
    save_cif_dir: str | None = None,
):
    two_theta_grid = np.linspace(two_theta_min, two_theta_max, n_points)
    X, y = [], []
    class_counts = {c: 0 for c in TARGET_SYSTEMS}

    cif_dir = None
    if save_cif_dir is not None:
        cif_dir = Path(save_cif_dir)
        cif_dir.mkdir(parents=True, exist_ok=True)

    xrd_calc = XRDCalculator(wavelength="CuKa")

    with MPRester(api_key) as mpr:
        robo_docs = mpr.materials.robocrys.search(keywords=["perovskite"])
        perov_mpids = [str(doc.material_id) for doc in robo_docs][:10000]

        summaries = mpr.materials.summary.search(
            material_ids=perov_mpids,
            fields=["material_id", "structure", "symmetry"],
        )

        for doc in summaries:
            sym = getattr(doc, "symmetry", None)
            if sym is None or sym.crystal_system is None:
                continue

            cs_raw = sym.crystal_system
            if hasattr(cs_raw, "value"):
                cs = str(cs_raw.value).lower()
            else:
                cs_str = str(cs_raw)
                if "." in cs_str:
                    cs_str = cs_str.split(".")[-1]
                cs = cs_str.lower()

            if cs not in TARGET_SYSTEMS:
                continue
            if class_counts[cs] >= max_per_class:
                continue

            structure = doc.structure
            if structure is None:
                continue

            if cif_dir is not None:
                mpid = str(doc.material_id)
                cif_str = structure.to(fmt="cif")
                (cif_dir / f"{mpid}_{cs}.cif").write_text(cif_str)

            try:
                pattern = xrd_calc.get_pattern(
                    structure,
                    two_theta_range=(two_theta_min, two_theta_max),
                )
                intensities = np.interp(two_theta_grid, pattern.x, pattern.y)
                X.append(intensities)
                y.append(cs)
                class_counts[cs] += 1
            except Exception as e:
                print(f"Error calculating XRD for {doc.material_id}: {e}")
                continue

            if all(class_counts[c] >= max_per_class for c in TARGET_SYSTEMS):
                break

    X = np.array(X, dtype=np.float32)
    y = np.array(y)

    print("Jumlah sampel per kelas:", class_counts)
    if cif_dir is not None:
        print(f"File CIF disimpan di folder: {cif_dir.resolve()}")

    return two_theta_grid, X, y


# ======================
# SAMPLING (TRAIN ONLY)
# ======================
def apply_sampling_train_only(X_train_raw, y_train, strategy_name):
    if strategy_name == "Original":
        return X_train_raw, y_train

    if strategy_name == "SMOTE":
        sampler = SMOTE(random_state=SEED, k_neighbors=5)
        return sampler.fit_resample(X_train_raw, y_train)

    if strategy_name == "ADASYN":
        sampler = ADASYN(random_state=SEED, n_neighbors=5)
        return sampler.fit_resample(X_train_raw, y_train)

    raise ValueError(f"Unknown strategy: {strategy_name}")


# ======================
# CNN FIXED
# ======================
def build_cnn_1d_fixed(input_shape, num_classes: int):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv1D(32, 7, activation="relu", padding="same")(inputs)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(64, 5, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(128, 3, activation="relu", padding="same")(x)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation="relu", name="feature_layer")(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model


# ======================
# FEATURES & METRICS
# ======================
def extract_cnn_features(model, X_data):
    if len(X_data.shape) == 2:
        X_data = X_data[..., np.newaxis]
    feature_layer = model.get_layer("feature_layer")
    feat_model = Model(inputs=model.input, outputs=feature_layer.output)
    return feat_model.predict(X_data, verbose=0, batch_size=32)


def calculate_separability_metrics(X_features, y):
    return {
        "silhouette_score": float(silhouette_score(X_features, y)),
        "davies_bouldin_index": float(davies_bouldin_score(X_features, y)),
        "calinski_harabasz_index": float(calinski_harabasz_score(X_features, y)),
    }


# ======================
# PLOTS
# ======================
def plot_class_distribution(y_encoded, class_names, title):
    unique, counts = np.unique(y_encoded, return_counts=True)
    labels = [class_names[i] for i in unique]
    plt.figure(figsize=(8, 5))
    bars = plt.bar(labels, counts)
    plt.title(title, fontweight="bold")
    plt.xticks(rotation=45)
    for b, c in zip(bars, counts):
        plt.text(b.get_x() + b.get_width()/2, b.get_height() + 10, str(c),
                 ha="center", va="bottom")
    plt.tight_layout()
    plt.show()


def plot_confusion_matrix(y_true, y_pred, class_names, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title, fontweight="bold")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))


def plot_umap_comparison_raw_vs_features(datasets_dict, class_names, save_prefix):
    n = len(datasets_dict)
    fig, axes = plt.subplots(n, 2, figsize=(16, 5*n))
    if n == 1:
        axes = axes.reshape(1, -1)

    colors = ["red", "blue", "green", "orange"]

    for idx, (strategy, data) in enumerate(datasets_dict.items()):
        X_raw = data["X_raw"]
        X_feat = data["X_features"]
        y = data["y"]

        # sample for speed
        n_samples = min(2000, len(X_raw))
        if len(X_raw) > n_samples:
            sel = np.random.choice(len(X_raw), n_samples, replace=False)
            X_raw_s = X_raw[sel]
            X_feat_s = X_feat[sel]
            y_s = y[sel]
        else:
            X_raw_s, X_feat_s, y_s = X_raw, X_feat, y

        umap_raw = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=SEED).fit_transform(X_raw_s)
        umap_feat = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=SEED).fit_transform(X_feat_s)

        ax1, ax2 = axes[idx, 0], axes[idx, 1]
        for i, lab in enumerate(np.unique(y_s)):
            m = (y_s == lab)
            name = class_names[lab]
            ax1.scatter(umap_raw[m, 0], umap_raw[m, 1], s=30, alpha=0.6,
                        c=colors[i % len(colors)], label=name, edgecolors="black", linewidths=0.2)
            ax2.scatter(umap_feat[m, 0], umap_feat[m, 1], s=30, alpha=0.6,
                        c=colors[i % len(colors)], label=name, edgecolors="black", linewidths=0.2)

        ax1.set_title(f"{strategy} - Raw (2048D→2D)", fontweight="bold")
        ax2.set_title(f"{strategy} - CNN Features (→2D)", fontweight="bold")
        ax1.grid(True, alpha=0.3); ax2.grid(True, alpha=0.3)
        ax1.legend(fontsize=8); ax2.legend(fontsize=8)

    plt.tight_layout()
    plt.savefig(f"{save_prefix}_raw_vs_features.png", dpi=150, bbox_inches="tight")
    plt.show()


# ======================
# TRAIN/EVAL PER STRATEGY
# ======================
def train_eval_cnn_only(X_raw, y_encoded, class_names, strategy_name):
    print("\n" + "="*70)
    print(f"[CNN ONLY] Strategy: {strategy_name}")
    print("="*70)

    # split first
    X_train_raw, X_test_raw, y_train, y_test = train_test_split(
        X_raw, y_encoded, test_size=0.2, random_state=SEED, stratify=y_encoded
    )

    # oversample only train
    X_train_res, y_train_res = apply_sampling_train_only(X_train_raw, y_train, strategy_name)

    print(f"Train before: {len(X_train_raw)} | after sampling: {len(X_train_res)}")
    print("Train dist:", Counter(y_train_res))
    print(f"Test untouched: {len(X_test_raw)} | dist:", Counter(y_test))

    # channel dim
    X_train = X_train_res[..., np.newaxis]
    X_test = X_test_raw[..., np.newaxis]

    model = build_cnn_1d_fixed(X_train.shape[1:], num_classes=len(class_names))
    model.summary()

    es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)

    history = model.fit(
        X_train, y_train_res,
        epochs=EPOCHS_FINAL,
        batch_size=BATCH_SIZE,
        validation_split=VAL_SPLIT,
        verbose=1,
        callbacks=[es]
    )

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    # features from all original X
    features_all = extract_cnn_features(model, X_raw[..., np.newaxis])
    sep = calculate_separability_metrics(features_all, y_encoded)

    print(f"Accuracy={acc:.4f} Precision={prec:.4f} Recall={rec:.4f} F1={f1:.4f}")
    print("Separability:", sep)

    return {
        "strategy": strategy_name,
        "history": history,
        "model": model,
        "y_test": y_test,
        "y_pred": y_pred,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "features": features_all,
        "separability": sep,
    }


# ======================
# MAIN
# ======================
if __name__ == "__main__":
    set_seed(SEED)

    print("="*60)
    print("CNN ONLY PIPELINE (Original / SMOTE / ADASYN)")
    print("="*60)

    two_theta, X_raw, y_str = load_perovskite_xrd_from_mp(API_KEY, save_cif_dir=str(CIF_SAVE_DIR))

    le = LabelEncoder()
    y_encoded = le.fit_transform(y_str)
    class_names = list(le.classes_)

    plot_class_distribution(y_encoded, class_names, "Original Class Distribution")

    strategies = ["Original", "SMOTE", "ADASYN"]

    results = {}
    datasets_for_umap = {}

    for strat in strategies:
        res = train_eval_cnn_only(X_raw, y_encoded, class_names, strat)
        results[strat] = res
        datasets_for_umap[strat] = {"X_raw": X_raw, "X_features": res["features"], "y": y_encoded}
        tf.keras.backend.clear_session()

    # UMAP
    plot_umap_comparison_raw_vs_features(datasets_for_umap, class_names, save_prefix="umap_CNN_only")

    # Metrics bar plot
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    metrics = ["accuracy", "precision", "recall", "f1"]
    titles = ["Accuracy", "Precision", "Recall", "F1-score"]
    for i, (m, t) in enumerate(zip(metrics, titles)):
        ax = axes[i//2, i % 2]
        vals = [results[s][m] for s in strategies]
        bars = ax.bar(strategies, vals, color=["gray", "green", "orange"])
        ax.set_title(f"{t} - CNN Only", fontweight="bold")
        ax.set_ylim(0, 1)
        ax.grid(True, alpha=0.3, axis="y")
        for b, v in zip(bars, vals):
            ax.text(b.get_x()+b.get_width()/2, b.get_height()+0.02, f"{v:.3f}", ha="center")
        plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha="right")

    plt.tight_layout()
    plt.savefig("metrics_CNN_only.png", dpi=150, bbox_inches="tight")
    plt.show()

    # Confusion matrices
    for strat in strategies:
        plot_confusion_matrix(results[strat]["y_test"], results[strat]["y_pred"], class_names,
                              title=f"Confusion Matrix - {strat} (CNN Only)")

    # Summary
    print("\n" + "="*70)
    print("SUMMARY (CNN ONLY)")
    print("="*70)
    print(f"{'Strategy':<10} {'Acc':<8} {'Prec':<8} {'Rec':<8} {'F1':<8} {'Sil':<10} {'DB':<10} {'CH':<10}")
    print("-"*80)
    for strat in strategies:
        r = results[strat]
        s = r["separability"]
        print(f"{strat:<10} {r['accuracy']:<8.4f} {r['precision']:<8.4f} {r['recall']:<8.4f} {r['f1']:<8.4f} "
              f"{s['silhouette_score']:<10.4f} {s['davies_bouldin_index']:<10.4f} {s['calinski_harabasz_index']:<10.2f}")

    print("\nGenerated files:")
    print(" - umap_CNN_only_raw_vs_features.png")
    print(" - metrics_CNN_only.png")
```

```python

# ============================================================
# woa_cnn_original_safe_umap.py
# - Strategy: Original
# - Data config: SAMA (N_POINTS=2048, MAX_PER_CLASS=5000, 4 kelas)
# - WOA+CNN lebih ringan: whales/epochs diperkecil, iters=3, bounds CNN diperkecil
# - UMAP tetap ada tetapi SAFE: tidak akan memicu import torch
#   Jika umap yang terinstall meng-import torch -> fallback PCA (tanpa crash)
# ============================================================

from __future__ import annotations

import os
import math
import json
import random
import importlib
import importlib.util
from pathlib import Path
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

# sesuai library lama (walau Original tidak pakai)
from imblearn.over_sampling import SMOTE, ADASYN

import tensorflow as tf
from tensorflow.keras import layers, models, Model, regularizers

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    silhouette_score, davies_bouldin_score, calinski_harabasz_score
)

# ======================
# CONFIG (DATA TETAP SAMA)
# ======================
API_KEY = "sMWxUUzNkgFJyEegqotWyVYFJEtK4N7K"  # <-- isi API key Materials Project kamu
TARGET_SYSTEMS = ["cubic", "monoclinic", "orthorhombic", "tetragonal"]

TWO_THETA_MIN = 10
TWO_THETA_MAX = 90
N_POINTS = 2048
MAX_PER_CLASS = 5000

HOME_DIR = Path.home()
CIF_SAVE_DIR = HOME_DIR / "Downloads" / "cif_perovskite"
RESULT_DIR = Path("results_woa_cnn_original_safe_umap")
RESULT_DIR.mkdir(exist_ok=True, parents=True)

SEED = 42

# Final training (tetap)
EPOCHS_FINAL = 30
BATCH_SIZE = 32

# ======================
# WOA lebih ringan (HANYA WOA/CNN yang diperingan)
# ======================
WOA_N_WHALES = 6
WOA_N_ITERS = 3
WOA_EPOCHS = 6
WOA_VAL_SIZE = 0.2

# Objective weights
LAMBDA_F1 = 0.7
LAMBDA_OVERLAP = 0.35

# overlap metric dipercepat via subsample fitur (tidak mengurangi data training)
OVERLAP_MAX_POINTS = 1200

# UMAP plot subsample (hanya untuk gambar)
UMAP_N_SAMPLES = 2000


# ======================
# SEED SETTER
# ======================
def set_seed(seed=SEED):
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)


# ======================
# SAFE UMAP (tanpa memicu import torch)
# ======================
def safe_umap_fit_transform(X, n_neighbors=15, min_dist=0.1, random_state=SEED):
    """
    Aman dari torch:
    - Tidak import 'umap' di top-level.
    - Cari spec 'umap' tanpa import.
    - Baca file origin (umap/__init__.py) bila ada. Jika mengandung 'import torch' -> stop.
    - Jika aman, import 'umap.umap_' dan gunakan UMAP.
    """
    # cek modul "umap" ada atau tidak, tanpa import
    spec = importlib.util.find_spec("umap")
    if spec is None:
        raise ImportError(
            "Module 'umap' tidak ditemukan. Install yang benar: 'umap-learn'."
        )

    # umap bisa namespace package -> origin None
    origin = getattr(spec, "origin", None)

    if origin:
        try:
            content = Path(origin).read_text(encoding="utf-8", errors="ignore")
        except Exception:
            content = ""
        if ("import torch" in content) or ("from torch" in content):
            raise ImportError(
                f"UMAP di environment ini ({origin}) meng-import torch -> akan crash.\n"
                "Uninstall paket 'umap' yang salah dan install 'umap-learn'."
            )

    # coba import umap.umap_ (umap-learn)
    umap_umap_ = importlib.import_module("umap.umap_")
    UMAP = getattr(umap_umap_, "UMAP", None)
    if UMAP is None:
        raise ImportError("Tidak menemukan umap.umap_.UMAP. Pastikan yang terinstal 'umap-learn'.")

    return UMAP(n_neighbors=n_neighbors, min_dist=min_dist, random_state=random_state).fit_transform(X)


def embed_2d(X, random_state=SEED):
    """
    Prioritas UMAP (safe). Jika gagal (karena umap salah/torch), fallback PCA.
    """
    try:
        return safe_umap_fit_transform(X, n_neighbors=15, min_dist=0.1, random_state=random_state)
    except Exception as e:
        print("\n[WARNING] UMAP tidak bisa dipakai tanpa memicu torch / tidak tersedia.")
        print("[WARNING] Alasan:", e)
        print("[WARNING] Fallback -> PCA (plot tetap ada, training tetap jalan)\n")
        return PCA(n_components=2, random_state=random_state).fit_transform(X)


# ======================
# label_smoothing FIX
# ======================
def make_sparse_ce_with_label_smoothing(num_classes: int, label_smoothing: float):
    label_smoothing = float(label_smoothing)
    if label_smoothing <= 0.0:
        return tf.keras.losses.SparseCategoricalCrossentropy()
    try:
        return tf.keras.losses.SparseCategoricalCrossentropy(label_smoothing=label_smoothing)
    except TypeError:
        def loss(y_true, y_pred):
            y_true = tf.cast(tf.reshape(y_true, [-1]), tf.int32)
            y_true_oh = tf.one_hot(y_true, depth=num_classes, dtype=y_pred.dtype)
            eps = tf.cast(label_smoothing, y_pred.dtype)
            smooth_pos = 1.0 - eps
            smooth_neg = eps / tf.cast(num_classes, y_pred.dtype)
            y_smooth = y_true_oh * smooth_pos + smooth_neg
            return tf.keras.losses.categorical_crossentropy(y_smooth, y_pred, from_logits=False)
        return loss


# ======================
# DATA LOADING (SAMA)
# ======================
def load_perovskite_xrd_from_mp(
    api_key: str,
    max_per_class: int = MAX_PER_CLASS,
    two_theta_min: float = TWO_THETA_MIN,
    two_theta_max: float = TWO_THETA_MAX,
    n_points: int = N_POINTS,
    save_cif_dir: str | None = None,
):
    two_theta_grid = np.linspace(two_theta_min, two_theta_max, n_points)
    X, y = [], []
    class_counts = {c: 0 for c in TARGET_SYSTEMS}

    cif_dir = None
    if save_cif_dir is not None:
        cif_dir = Path(save_cif_dir)
        cif_dir.mkdir(parents=True, exist_ok=True)

    xrd_calc = XRDCalculator(wavelength="CuKa")

    with MPRester(api_key) as mpr:
        robo_docs = mpr.materials.robocrys.search(keywords=["perovskite"])
        perov_mpids = [str(doc.material_id) for doc in robo_docs][:10000]

        summaries = mpr.materials.summary.search(
            material_ids=perov_mpids,
            fields=["material_id", "structure", "symmetry"],
        )

        for doc in summaries:
            sym = getattr(doc, "symmetry", None)
            if sym is None or sym.crystal_system is None:
                continue

            cs_raw = sym.crystal_system
            if hasattr(cs_raw, "value"):
                cs = str(cs_raw.value).lower()
            else:
                cs_str = str(cs_raw)
                if "." in cs_str:
                    cs_str = cs_str.split(".")[-1]
                cs = cs_str.lower()

            if cs not in TARGET_SYSTEMS:
                continue
            if class_counts[cs] >= max_per_class:
                continue

            structure = doc.structure
            if structure is None:
                continue

            if cif_dir is not None:
                mpid = str(doc.material_id)
                (cif_dir / f"{mpid}_{cs}.cif").write_text(structure.to(fmt="cif"))

            try:
                pattern = xrd_calc.get_pattern(
                    structure,
                    two_theta_range=(two_theta_min, two_theta_max),
                )
                intensities = np.interp(two_theta_grid, pattern.x, pattern.y)
                X.append(intensities)
                y.append(cs)
                class_counts[cs] += 1
            except Exception as e:
                print(f"Error calculating XRD for {doc.material_id}: {e}")
                continue

            if all(class_counts[c] >= max_per_class for c in TARGET_SYSTEMS):
                break

    X = np.array(X, dtype=np.float32)
    y = np.array(y)

    print("Jumlah sampel per kelas:", class_counts)
    if cif_dir is not None:
        print(f"File CIF disimpan di folder: {cif_dir.resolve()}")

    return two_theta_grid, X, y


# ======================
# RESAMPLING (Original = no-op)
# ======================
def resample_train(X_train_raw, y_train, strategy_name, k_neighbors=5):
    return X_train_raw, y_train


# ======================
# CNN
# ======================
def build_cnn_1d_hp(input_shape, num_classes, hp: dict):
    l2_reg = float(hp.get("l2_reg", 0.0))
    bn = int(hp.get("batch_norm", 1))
    dr = float(hp.get("dropout", 0.0))
    reg = regularizers.l2(l2_reg) if l2_reg > 0 else None

    inputs = layers.Input(shape=input_shape)

    x = layers.Conv1D(int(hp["f1"]), int(hp["k1"]), padding="same", kernel_regularizer=reg)(inputs)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f2"]), int(hp["k2"]), padding="same", kernel_regularizer=reg)(x)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f3"]), int(hp["k3"]), padding="same", kernel_regularizer=reg)(x)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.GlobalAveragePooling1D()(x)

    x = layers.Dense(int(hp["dense"]), activation="relu", kernel_regularizer=reg, name="feature_layer")(x)
    if dr > 0:
        x = layers.Dropout(dr)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = models.Model(inputs, outputs)

    loss_fn = make_sparse_ce_with_label_smoothing(num_classes, float(hp.get("label_smoothing", 0.0)))
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=float(hp["lr"])),
        loss=loss_fn,
        metrics=["accuracy"],
    )
    return model


# ======================
# FEATURE + OVERLAP
# ======================
def get_feature_model(model):
    return Model(inputs=model.input, outputs=model.get_layer("feature_layer").output)


def safe_separability_metrics(features, y):
    sil = dbi = chi = None
    try:
        if len(np.unique(y)) >= 2 and features.shape[0] > len(np.unique(y)):
            sil = float(silhouette_score(features, y))
    except Exception:
        pass
    try:
        if len(np.unique(y)) >= 2:
            dbi = float(davies_bouldin_score(features, y))
    except Exception:
        pass
    try:
        if len(np.unique(y)) >= 2:
            chi = float(calinski_harabasz_score(features, y))
    except Exception:
        pass
    return sil, dbi, chi


def overlap_penalty_from_features(features, y):
    n = features.shape[0]
    if n > OVERLAP_MAX_POINTS:
        idx = np.random.choice(n, OVERLAP_MAX_POINTS, replace=False)
        features = features[idx]
        y = y[idx]

    sil, dbi, _ = safe_separability_metrics(features, y)

    sil_pen = 1.0 if sil is None else (1.0 - ((sil + 1.0) / 2.0))
    dbi_pen = 1.0 if dbi is None else float(np.log1p(dbi))
    return sil_pen + dbi_pen


# ======================
# WOA
# ======================
def decode_whale_position(pos, bounds):
    hp = {}
    for i, (name, (low, high, t)) in enumerate(bounds):
        x = pos[i]
        if t == "int":
            hp[name] = int(round(low + x * (high - low)))
        elif t == "float":
            hp[name] = float(low + x * (high - low))
        elif t == "float_log":
            hp[name] = 10 ** (math.log10(low) + x * (math.log10(high) - math.log10(low)))
        elif t == "bool":
            hp[name] = int(round(x))
        else:
            raise ValueError(f"Unknown type: {t}")

    for k in ["k1", "k2", "k3"]:
        if hp[k] % 2 == 0:
            hp[k] += 1
        hp[k] = int(np.clip(hp[k], 3, 17))

    hp["dropout"] = float(np.clip(hp.get("dropout", 0.0), 0.0, 0.6))
    hp["label_smoothing"] = float(np.clip(hp.get("label_smoothing", 0.0), 0.0, 0.2))
    hp["l2_reg"] = float(np.clip(hp.get("l2_reg", 0.0), 0.0, 1e-2))
    return hp


def woa_optimize_original(X_train_raw, y_train, input_shape, num_classes):
    # bounds diperkecil agar kandidat CNN lebih ringan
    bounds = [
        ("lr", (5e-5, 2e-3, "float_log")),
        ("f1", (8, 48, "int")),
        ("k1", (3, 11, "int")),
        ("f2", (16, 96, "int")),
        ("k2", (3, 9, "int")),
        ("f3", (32, 128, "int")),
        ("k3", (3, 7, "int")),
        ("dense", (32, 160, "int")),
        ("dropout", (0.0, 0.5, "float")),
        ("l2_reg", (0.0, 8e-3, "float")),
        ("batch_norm", (0, 1, "bool")),
        ("label_smoothing", (0.0, 0.2, "float")),
    ]

    dim = len(bounds)
    whales = np.random.rand(WOA_N_WHALES, dim)
    best_score, best_hp, best_pos = float("inf"), None, whales[0].copy()

    def fitness(pos):
        hp = decode_whale_position(pos, bounds)

        X_res, y_res = resample_train(X_train_raw, y_train, "Original")
        X_tr, X_val, y_tr, y_val = train_test_split(
            X_res, y_res, test_size=WOA_VAL_SIZE, random_state=SEED, stratify=y_res
        )

        X_tr_c, X_val_c = X_tr[..., np.newaxis], X_val[..., np.newaxis]
        model = build_cnn_1d_hp(input_shape, num_classes, hp)

        es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=1, restore_best_weights=True)
        hist = model.fit(
            X_tr_c, y_tr,
            epochs=WOA_EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_val_c, y_val),
            verbose=0,
            callbacks=[es],
        )

        val_loss = float(np.min(hist.history["val_loss"]))
        y_val_pred = np.argmax(model.predict(X_val_c, verbose=0), axis=1)
        val_macro_f1 = float(f1_score(y_val, y_val_pred, average="macro", zero_division=0))

        feat_model = get_feature_model(model)
        val_features = feat_model.predict(X_val_c, verbose=0, batch_size=128)
        overlap_pen = float(overlap_penalty_from_features(val_features, y_val))

        score = val_loss + LAMBDA_F1 * (1.0 - val_macro_f1) + LAMBDA_OVERLAP * overlap_pen

        tf.keras.backend.clear_session()
        return score, hp

    # initial
    for i in range(WOA_N_WHALES):
        sc, hp = fitness(whales[i])
        if sc < best_score:
            best_score, best_hp, best_pos = sc, hp, whales[i].copy()

    # iter (3)
    for t in range(WOA_N_ITERS):
        a = 2 - t * (2 / WOA_N_ITERS)
        for i in range(WOA_N_WHALES):
            r1, r2 = np.random.rand(dim), np.random.rand(dim)
            A = 2 * a * r1 - a
            C = 2 * r2
            p = np.random.rand()

            if p < 0.5:
                if np.linalg.norm(A) < 1:
                    D = np.abs(C * best_pos - whales[i])
                    new_pos = best_pos - A * D
                else:
                    X_rand = whales[np.random.randint(WOA_N_WHALES)]
                    D = np.abs(C * X_rand - whales[i])
                    new_pos = X_rand - A * D
            else:
                D = np.abs(best_pos - whales[i])
                b = 1
                l = np.random.uniform(-1, 1, dim)
                new_pos = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_pos

            whales[i] = np.clip(new_pos, 0, 1)

        for i in range(WOA_N_WHALES):
            sc, hp = fitness(whales[i])
            if sc < best_score:
                best_score, best_hp, best_pos = sc, hp, whales[i].copy()

        print(f"[WOA-Original] iter {t+1}/{WOA_N_ITERS} best_score={best_score:.5f}")
        print("  best_hp:", best_hp)

    return best_hp, best_score


# ======================
# PLOTS
# ======================
def plot_confusion_matrix(y_true, y_pred, class_names, title, save_path=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title, fontweight="bold")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))


def plot_umap_raw_vs_features_single(X_raw, features, y_encoded, class_names, title_prefix, save_path):
    n = len(X_raw)
    n_s = min(UMAP_N_SAMPLES, n)
    if n > n_s:
        idx = np.random.choice(n, n_s, replace=False)
        X_s = X_raw[idx]
        feats_s = features[idx]
        y_s = y_encoded[idx]
    else:
        X_s, feats_s, y_s = X_raw, features, y_encoded

    emb_raw = embed_2d(X_s, random_state=SEED)
    emb_feat = embed_2d(feats_s, random_state=SEED)

    colors = ["red", "blue", "green", "orange", "purple", "brown"]
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    for i, lab in enumerate(np.unique(y_s)):
        m = y_s == lab
        axes[0].scatter(emb_raw[m, 0], emb_raw[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)
        axes[1].scatter(emb_feat[m, 0], emb_feat[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)

    axes[0].set_title(f"{title_prefix} - Raw (2048D→2D)", fontweight="bold")
    axes[1].set_title(f"{title_prefix} - Features (→2D)", fontweight="bold")
    axes[0].grid(True, alpha=0.3)
    axes[1].grid(True, alpha=0.3)
    axes[0].legend(fontsize=8)
    axes[1].legend(fontsize=8)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()


# ======================
# MAIN
# ======================
if __name__ == "__main__":
    set_seed(SEED)

    print("=" * 70)
    print("WOA + CNN (Original) - SAFE UMAP (tanpa import torch)")
    print("=" * 70)

    _, X_raw, y_str = load_perovskite_xrd_from_mp(API_KEY, save_cif_dir=str(CIF_SAVE_DIR))

    le = LabelEncoder()
    y_encoded = le.fit_transform(y_str)
    class_names = list(le.classes_)

    X_train_raw, X_test_raw, y_train, y_test = train_test_split(
        X_raw, y_encoded, test_size=0.2, random_state=SEED, stratify=y_encoded
    )

    print(f"Train raw: {len(X_train_raw)} | Test raw: {len(X_test_raw)}")
    print("Train dist(raw):", Counter(y_train))
    print("Test dist:", Counter(y_test))

    input_shape = (X_train_raw.shape[1], 1)
    num_classes = len(class_names)

    best_hp, best_score = woa_optimize_original(X_train_raw, y_train, input_shape, num_classes)

    print("\n=== BEST FOUND (WOA) ===")
    print("Best score:", best_score)
    print("Best hp:", best_hp)

    # Final training
    X_train = X_train_raw[..., np.newaxis]
    X_test = X_test_raw[..., np.newaxis]

    model = build_cnn_1d_hp(input_shape, num_classes, best_hp)
    es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        epochs=EPOCHS_FINAL,
        batch_size=BATCH_SIZE,
        validation_split=0.2,
        verbose=1,
        callbacks=[es]
    )

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    print("\n=== TEST RESULTS ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")

    # Features for embedding plot
    feat_model = get_feature_model(model)
    features_all = feat_model.predict(X_raw[..., np.newaxis], verbose=0, batch_size=128)

    # Save best hp
    with open(RESULT_DIR / "best_hp.json", "w", encoding="utf-8") as f:
        json.dump(best_hp, f, indent=2)

    plot_confusion_matrix(
        y_test, y_pred, class_names,
        title="Confusion Matrix - Original (WOA+CNN)",
        save_path=str(RESULT_DIR / "cm.png")
    )

    plot_umap_raw_vs_features_single(
        X_raw=X_raw,
        features=features_all,
        y_encoded=y_encoded,
        class_names=class_names,
        title_prefix="Original",
        save_path=str(RESULT_DIR / "umap_raw_vs_features.png")
    )

    tf.keras.backend.clear_session()
    print("\nSaved outputs in:", RESULT_DIR.resolve())
    print("DONE.")
```

```python

# ============================================================
# woa_cnn_smote.py (Standalone - LIGHT WOA/CNN + SAFE UMAP)
# - Strategy: SMOTE (TRAIN ONLY)
# - DATA: sama (N_POINTS=2048, MAX_PER_CLASS=5000)
# - UMAP: tetap ada, safe import (tidak memicu torch)
# ============================================================

from __future__ import annotations

import os
import math
import json
import random
from pathlib import Path
from collections import Counter
import importlib.util
import importlib

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

from imblearn.over_sampling import SMOTE, ADASYN

import tensorflow as tf
from tensorflow.keras import layers, models, Model, regularizers

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    silhouette_score, davies_bouldin_score, calinski_harabasz_score
)

# ======================
# CONFIG (DATA TETAP SAMA)
# ======================
API_KEY = "sMWxUUzNkgFJyEegqotWyVYFJEtK4N7K"  # <-- isi API key kamu
TARGET_SYSTEMS = ["cubic", "monoclinic", "orthorhombic", "tetragonal"]

TWO_THETA_MIN = 10
TWO_THETA_MAX = 90
N_POINTS = 2048
MAX_PER_CLASS = 5000

HOME_DIR = Path.home()
CIF_SAVE_DIR = HOME_DIR / "Downloads" / "cif_perovskite"
RESULT_DIR = Path("results_woa_cnn_smote")
RESULT_DIR.mkdir(exist_ok=True, parents=True)

SEED = 42

EPOCHS_FINAL = 30
BATCH_SIZE = 32

# WOA ringan
WOA_N_WHALES = 6
WOA_N_ITERS = 3
WOA_EPOCHS = 6
WOA_VAL_SIZE = 0.2

LAMBDA_F1 = 0.7
LAMBDA_OVERLAP = 0.35

OVERLAP_MAX_POINTS = 1200

# UMAP (plot subsample)
UMAP_N_SAMPLES = 2000
UMAP_N_NEIGHBORS = 15
UMAP_MIN_DIST = 0.1


def set_seed(seed=SEED):
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)


# ======================
# SAFE UMAP (NO TORCH)
# ======================
def safe_umap_fit_transform(X, n_neighbors=15, min_dist=0.1, random_state=42):
    spec = importlib.util.find_spec("umap")
    if spec is None or spec.origin is None:
        raise ImportError("Module 'umap' tidak ditemukan. Install 'umap-learn'.")

    init_path = Path(spec.origin)

    content = ""
    try:
        content = init_path.read_text(encoding="utf-8", errors="ignore")
    except Exception:
        content = ""

    if ("import torch" in content) or ("from torch" in content):
        raise ImportError(f"UMAP terdeteksi mengimpor torch ({init_path}). UMAP diblok demi keamanan.")

    umap_umap_ = importlib.import_module("umap.umap_")
    UMAP = getattr(umap_umap_, "UMAP", None)
    if UMAP is None:
        raise ImportError("Tidak menemukan umap.umap_.UMAP. Pastikan install 'umap-learn'.")

    return UMAP(n_neighbors=n_neighbors, min_dist=min_dist, random_state=random_state).fit_transform(X)


def embed_2d(X, random_state=SEED):
    try:
        return safe_umap_fit_transform(X, UMAP_N_NEIGHBORS, UMAP_MIN_DIST, random_state)
    except Exception as e:
        print("\n[WARNING] UMAP tidak dapat digunakan (tanpa memicu torch). Fallback -> PCA.")
        print("[WARNING] Alasan:", e, "\n")
        return PCA(n_components=2, random_state=random_state).fit_transform(X)


def make_sparse_ce_with_label_smoothing(num_classes: int, label_smoothing: float):
    label_smoothing = float(label_smoothing)
    if label_smoothing <= 0.0:
        return tf.keras.losses.SparseCategoricalCrossentropy()
    try:
        return tf.keras.losses.SparseCategoricalCrossentropy(label_smoothing=label_smoothing)
    except TypeError:
        def loss(y_true, y_pred):
            y_true = tf.cast(tf.reshape(y_true, [-1]), tf.int32)
            y_true_oh = tf.one_hot(y_true, depth=num_classes, dtype=y_pred.dtype)
            eps = tf.cast(label_smoothing, y_pred.dtype)
            y_smooth = y_true_oh * (1.0 - eps) + eps / tf.cast(num_classes, y_pred.dtype)
            return tf.keras.losses.categorical_crossentropy(y_smooth, y_pred, from_logits=False)
        return loss


def load_perovskite_xrd_from_mp(
    api_key: str,
    max_per_class: int = MAX_PER_CLASS,
    two_theta_min: float = TWO_THETA_MIN,
    two_theta_max: float = TWO_THETA_MAX,
    n_points: int = N_POINTS,
    save_cif_dir: str | None = None,
):
    two_theta_grid = np.linspace(two_theta_min, two_theta_max, n_points)
    X, y = [], []
    class_counts = {c: 0 for c in TARGET_SYSTEMS}

    cif_dir = None
    if save_cif_dir is not None:
        cif_dir = Path(save_cif_dir)
        cif_dir.mkdir(parents=True, exist_ok=True)

    xrd_calc = XRDCalculator(wavelength="CuKa")

    with MPRester(api_key) as mpr:
        robo_docs = mpr.materials.robocrys.search(keywords=["perovskite"])
        perov_mpids = [str(doc.material_id) for doc in robo_docs][:10000]

        summaries = mpr.materials.summary.search(
            material_ids=perov_mpids,
            fields=["material_id", "structure", "symmetry"],
        )

        for doc in summaries:
            sym = getattr(doc, "symmetry", None)
            if sym is None or sym.crystal_system is None:
                continue

            cs_raw = sym.crystal_system
            if hasattr(cs_raw, "value"):
                cs = str(cs_raw.value).lower()
            else:
                cs = str(cs_raw).split(".")[-1].lower()

            if cs not in TARGET_SYSTEMS:
                continue
            if class_counts[cs] >= max_per_class:
                continue

            structure = doc.structure
            if structure is None:
                continue

            if cif_dir is not None:
                mpid = str(doc.material_id)
                (cif_dir / f"{mpid}_{cs}.cif").write_text(structure.to(fmt="cif"))

            try:
                pattern = xrd_calc.get_pattern(structure, two_theta_range=(two_theta_min, two_theta_max))
                intensities = np.interp(two_theta_grid, pattern.x, pattern.y)
                X.append(intensities)
                y.append(cs)
                class_counts[cs] += 1
            except Exception as e:
                print(f"Error XRD for {doc.material_id}: {e}")
                continue

            if all(class_counts[c] >= max_per_class for c in TARGET_SYSTEMS):
                break

    X = np.array(X, dtype=np.float32)
    y = np.array(y)
    print("Jumlah sampel per kelas:", class_counts)
    return two_theta_grid, X, y


def resample_train(X_train_raw, y_train, k_neighbors=5):
    k_neighbors = int(np.clip(k_neighbors, 3, 12))
    sampler = SMOTE(random_state=SEED, k_neighbors=k_neighbors)
    return sampler.fit_resample(X_train_raw, y_train)


def build_cnn_1d_hp(input_shape, num_classes, hp: dict):
    l2_reg = float(hp.get("l2_reg", 0.0))
    bn = int(hp.get("batch_norm", 1))
    dr = float(hp.get("dropout", 0.0))
    reg = regularizers.l2(l2_reg) if l2_reg > 0 else None

    inputs = layers.Input(shape=input_shape)

    x = layers.Conv1D(int(hp["f1"]), int(hp["k1"]), padding="same", kernel_regularizer=reg)(inputs)
    if bn: x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f2"]), int(hp["k2"]), padding="same", kernel_regularizer=reg)(x)
    if bn: x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f3"]), int(hp["k3"]), padding="same", kernel_regularizer=reg)(x)
    if bn: x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(int(hp["dense"]), activation="relu", kernel_regularizer=reg, name="feature_layer")(x)
    if dr > 0:
        x = layers.Dropout(dr)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = models.Model(inputs, outputs)

    loss_fn = make_sparse_ce_with_label_smoothing(num_classes, float(hp.get("label_smoothing", 0.0)))
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=float(hp["lr"])),
        loss=loss_fn,
        metrics=["accuracy"],
    )
    return model


def get_feature_model(model):
    return Model(inputs=model.input, outputs=model.get_layer("feature_layer").output)


def safe_separability_metrics(features, y):
    sil = dbi = chi = None
    try:
        if len(np.unique(y)) >= 2 and features.shape[0] > len(np.unique(y)):
            sil = float(silhouette_score(features, y))
    except Exception:
        pass
    try:
        if len(np.unique(y)) >= 2:
            dbi = float(davies_bouldin_score(features, y))
    except Exception:
        pass
    try:
        if len(np.unique(y)) >= 2:
            chi = float(calinski_harabasz_score(features, y))
    except Exception:
        pass
    return sil, dbi, chi


def overlap_penalty_from_features(features, y):
    n = features.shape[0]
    if n > OVERLAP_MAX_POINTS:
        idx = np.random.choice(n, OVERLAP_MAX_POINTS, replace=False)
        features = features[idx]
        y = y[idx]

    sil, dbi, _ = safe_separability_metrics(features, y)
    sil_pen = 1.0 if sil is None else (1.0 - ((sil + 1.0) / 2.0))
    dbi_pen = 1.0 if dbi is None else float(np.log1p(dbi))
    return sil_pen + dbi_pen


def decode_whale_position(pos, bounds):
    hp = {}
    for i, (name, (low, high, t)) in enumerate(bounds):
        x = pos[i]
        if t == "int":
            hp[name] = int(round(low + x * (high - low)))
        elif t == "float":
            hp[name] = float(low + x * (high - low))
        elif t == "float_log":
            hp[name] = 10 ** (math.log10(low) + x * (math.log10(high) - math.log10(low)))
        elif t == "bool":
            hp[name] = int(round(x))
        else:
            raise ValueError(t)

    for k in ["k1", "k2", "k3"]:
        if hp[k] % 2 == 0:
            hp[k] += 1
        hp[k] = int(np.clip(hp[k], 3, 17))

    hp["dropout"] = float(np.clip(hp.get("dropout", 0.0), 0.0, 0.6))
    hp["label_smoothing"] = float(np.clip(hp.get("label_smoothing", 0.0), 0.0, 0.2))
    hp["l2_reg"] = float(np.clip(hp.get("l2_reg", 0.0), 0.0, 1e-2))
    hp["k_neighbors"] = int(np.clip(hp.get("k_neighbors", 5), 3, 12))
    return hp


def woa_optimize_for_smote(X_train_raw, y_train, input_shape, num_classes):
    bounds = [
        ("lr", (5e-5, 2e-3, "float_log")),
        ("f1", (8, 48, "int")),
        ("k1", (3, 11, "int")),
        ("f2", (16, 96, "int")),
        ("k2", (3, 9, "int")),
        ("f3", (32, 128, "int")),
        ("k3", (3, 7, "int")),
        ("dense", (32, 160, "int")),
        ("dropout", (0.0, 0.5, "float")),
        ("l2_reg", (0.0, 8e-3, "float")),
        ("batch_norm", (0, 1, "bool")),
        ("label_smoothing", (0.0, 0.2, "float")),
        ("k_neighbors", (3, 12, "int")),
    ]

    dim = len(bounds)
    whales = np.random.rand(WOA_N_WHALES, dim)

    best_score = float("inf")
    best_hp = None
    best_pos = whales[0].copy()

    def fitness(pos):
        hp = decode_whale_position(pos, bounds)

        X_res, y_res = resample_train(X_train_raw, y_train, k_neighbors=hp["k_neighbors"])
        X_tr, X_val, y_tr, y_val = train_test_split(
            X_res, y_res, test_size=WOA_VAL_SIZE, random_state=SEED, stratify=y_res
        )

        X_tr_c = X_tr[..., np.newaxis]
        X_val_c = X_val[..., np.newaxis]

        model = build_cnn_1d_hp(input_shape, num_classes, hp)
        es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=1, restore_best_weights=True)

        hist = model.fit(
            X_tr_c, y_tr,
            epochs=WOA_EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_val_c, y_val),
            verbose=0,
            callbacks=[es],
        )

        val_loss = float(np.min(hist.history["val_loss"]))
        y_val_pred = np.argmax(model.predict(X_val_c, verbose=0), axis=1)
        val_macro_f1 = float(f1_score(y_val, y_val_pred, average="macro", zero_division=0))

        feat_model = get_feature_model(model)
        val_features = feat_model.predict(X_val_c, verbose=0, batch_size=128)
        overlap_pen = float(overlap_penalty_from_features(val_features, y_val))

        score = val_loss + LAMBDA_F1 * (1.0 - val_macro_f1) + LAMBDA_OVERLAP * overlap_pen

        tf.keras.backend.clear_session()
        return score, hp

    for i in range(WOA_N_WHALES):
        sc, hp = fitness(whales[i])
        if sc < best_score:
            best_score, best_hp, best_pos = sc, hp, whales[i].copy()

    for t in range(WOA_N_ITERS):
        a = 2 - t * (2 / WOA_N_ITERS)

        for i in range(WOA_N_WHALES):
            r1 = np.random.rand(dim)
            r2 = np.random.rand(dim)
            A = 2 * a * r1 - a
            C = 2 * r2
            p = np.random.rand()

            if p < 0.5:
                if np.linalg.norm(A) < 1:
                    D = np.abs(C * best_pos - whales[i])
                    new_pos = best_pos - A * D
                else:
                    rand_idx = np.random.randint(WOA_N_WHALES)
                    X_rand = whales[rand_idx]
                    D = np.abs(C * X_rand - whales[i])
                    new_pos = X_rand - A * D
            else:
                D = np.abs(best_pos - whales[i])
                b = 1
                l = np.random.uniform(-1, 1, dim)
                new_pos = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_pos

            whales[i] = np.clip(new_pos, 0, 1)

        for i in range(WOA_N_WHALES):
            sc, hp = fitness(whales[i])
            if sc < best_score:
                best_score, best_hp, best_pos = sc, hp, whales[i].copy()

        print(f"[WOA-SMOTE] iter {t+1}/{WOA_N_ITERS} best_score={best_score:.5f}")
        print("  best_hp:", best_hp)

    return best_hp, best_score


def plot_confusion_matrix(y_true, y_pred, class_names, title, save_path=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title, fontweight="bold")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))


def plot_umap_raw_vs_features_single(X_raw, features, y_encoded, class_names, title_prefix, save_path):
    n = len(X_raw)
    n_s = min(UMAP_N_SAMPLES, n)
    if n > n_s:
        idx = np.random.choice(n, n_s, replace=False)
        X_s = X_raw[idx]
        feats_s = features[idx]
        y_s = y_encoded[idx]
    else:
        X_s, feats_s, y_s = X_raw, features, y_encoded

    emb_raw = embed_2d(X_s, random_state=SEED)
    emb_feat = embed_2d(feats_s, random_state=SEED)

    colors = ["red", "blue", "green", "orange", "purple", "brown"]
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    for i, lab in enumerate(np.unique(y_s)):
        m = y_s == lab
        axes[0].scatter(emb_raw[m, 0], emb_raw[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)
        axes[1].scatter(emb_feat[m, 0], emb_feat[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)

    axes[0].set_title(f"{title_prefix} - Raw (2048D→2D)", fontweight="bold")
    axes[1].set_title(f"{title_prefix} - Features (→2D)", fontweight="bold")
    axes[0].grid(True, alpha=0.3)
    axes[1].grid(True, alpha=0.3)
    axes[0].legend(fontsize=8)
    axes[1].legend(fontsize=8)

    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    set_seed(SEED)

    print("=" * 60)
    print("WOA + CNN (SMOTE) - LIGHT + SAFE UMAP (NO TORCH IMPORT)")
    print("=" * 60)

    _, X_raw, y_str = load_perovskite_xrd_from_mp(API_KEY, save_cif_dir=str(CIF_SAVE_DIR))

    le = LabelEncoder()
    y_encoded = le.fit_transform(y_str)
    class_names = list(le.classes_)

    X_train_raw, X_test_raw, y_train, y_test = train_test_split(
        X_raw, y_encoded, test_size=0.2, random_state=SEED, stratify=y_encoded
    )

    input_shape = (X_train_raw.shape[1], 1)
    num_classes = len(class_names)

    best_hp, best_score = woa_optimize_for_smote(X_train_raw, y_train, input_shape, num_classes)
    print("\n=== BEST FOUND (WOA) ===")
    print("Best score:", best_score)
    print("Best hp:", best_hp)

    X_train_res, y_train_res = resample_train(X_train_raw, y_train, k_neighbors=best_hp["k_neighbors"])

    X_train = X_train_res[..., np.newaxis]
    X_test = X_test_raw[..., np.newaxis]

    model = build_cnn_1d_hp(input_shape, num_classes, best_hp)
    es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)

    history = model.fit(
        X_train, y_train_res,
        epochs=EPOCHS_FINAL,
        batch_size=BATCH_SIZE,
        validation_split=0.2,
        verbose=1,
        callbacks=[es],
    )

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    print("\n=== TEST RESULTS ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")

    feat_model = get_feature_model(model)
    features_all = feat_model.predict(X_raw[..., np.newaxis], verbose=0, batch_size=128)

    with open(RESULT_DIR / "best_hp.json", "w", encoding="utf-8") as f:
        json.dump(best_hp, f, indent=2)

    plot_confusion_matrix(
        y_test, y_pred, class_names,
        title="Confusion Matrix - SMOTE (WOA+CNN)",
        save_path=str(RESULT_DIR / "cm.png"),
    )

    plot_umap_raw_vs_features_single(
        X_raw=X_raw,
        features=features_all,
        y_encoded=y_encoded,
        class_names=class_names,
        title_prefix="SMOTE",
        save_path=str(RESULT_DIR / "umap_raw_vs_features.png"),
    )

    tf.keras.backend.clear_session()
    print("\nSaved outputs in:", RESULT_DIR.resolve())
    print("DONE.")
```

```python

# ============================================================
# woa_cnn_adasyn_revised_light_no_torch.py
# Standalone WOA + CNN for ADASYN
#
# - DATA TETAP SAMA:
#   TARGET_SYSTEMS, N_POINTS=2048, MAX_PER_CLASS=5000, TWO_THETA range sama
#
# - WOA + CNN DIPERINGAN:
#   WOA_N_ITERS=3, WOA_N_WHALES=6, WOA_EPOCHS=6
#   bounds CNN diperkecil (kandidat cepat)
#   early stopping WOA agresif (patience=1)
#   overlap penalty dihitung dengan subsample feature (untuk fitness saja)
#
# - UMAP TETAP ADA tapi TIDAK MEMICU TORCH:
#   Tidak ada "import umap" di top-level.
#   safe_umap_fit_transform() cek isi umap/__init__.py:
#     jika ada import torch -> TIDAK import umap, fallback ke PCA
# ============================================================

from __future__ import annotations

import os
import math
import json
import random
import importlib
import importlib.util
from pathlib import Path
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

from imblearn.over_sampling import ADASYN

import tensorflow as tf
from tensorflow.keras import layers, models, Model, regularizers

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix,
    silhouette_score, davies_bouldin_score, calinski_harabasz_score
)
from sklearn.decomposition import PCA


# ======================
# CONFIG (DATA TETAP SAMA)
# ======================
API_KEY = "sMWxUUzNkgFJyEegqotWyVYFJEtK4N7K"  # <-- isi API key Materials Project kamu
STRATEGY_NAME = "ADASYN"

TARGET_SYSTEMS = ["cubic", "monoclinic", "orthorhombic", "tetragonal"]

TWO_THETA_MIN = 10
TWO_THETA_MAX = 90
N_POINTS = 2048
MAX_PER_CLASS = 5000

HOME_DIR = Path.home()
CIF_SAVE_DIR = HOME_DIR / "Downloads" / "cif_perovskite"

RESULT_DIR = Path("results_woa_cnn_adasyn_revised")
RESULT_DIR.mkdir(exist_ok=True, parents=True)

SEED = 42

# Final training (biarkan sama seperti awal)
EPOCHS_FINAL = 30
BATCH_SIZE = 32

# ======================
# WOA CONFIG (DIPERINGAN)
# ======================
WOA_N_WHALES = 6
WOA_N_ITERS = 3
WOA_EPOCHS = 6
WOA_VAL_SIZE = 0.2

# Objective weights
LAMBDA_F1 = 0.7
LAMBDA_OVERLAP = 0.35

# overlap penalty dihitung subsample (untuk fitness WOA saja)
OVERLAP_MAX_POINTS = 1200

# UMAP plot subsample (hanya visualisasi)
UMAP_N_SAMPLES = 2000


# ======================
# SEED SETTER
# ======================
def set_seed(seed=SEED):
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)


# ======================
# label_smoothing FIX (robust loss factory)
# ======================
def make_sparse_ce_with_label_smoothing(num_classes: int, label_smoothing: float):
    """
    Robust loss factory:
    - If TF supports label_smoothing in SparseCategoricalCrossentropy -> use it.
    - Else fallback to custom sparse label smoothing (sparse -> one-hot -> smooth -> CCE).
    """
    label_smoothing = float(label_smoothing)

    if label_smoothing <= 0.0:
        return tf.keras.losses.SparseCategoricalCrossentropy()

    try:
        return tf.keras.losses.SparseCategoricalCrossentropy(label_smoothing=label_smoothing)
    except TypeError:
        def loss(y_true, y_pred):
            y_true = tf.cast(tf.reshape(y_true, [-1]), tf.int32)
            y_true_oh = tf.one_hot(y_true, depth=num_classes, dtype=y_pred.dtype)

            eps = tf.cast(label_smoothing, y_pred.dtype)
            smooth_pos = 1.0 - eps
            smooth_neg = eps / tf.cast(num_classes, y_pred.dtype)

            y_smooth = y_true_oh * smooth_pos + smooth_neg
            return tf.keras.losses.categorical_crossentropy(y_smooth, y_pred, from_logits=False)

        return loss


# ======================
# SAFE UMAP LOADER (TIDAK MEMICU TORCH)
# ======================
def safe_umap_fit_transform(X, n_neighbors=15, min_dist=0.1, random_state=42):
    """
    Aman dari torch:
    - Tidak import 'umap' dulu.
    - Cek file umap/__init__.py, jika ada 'import torch' -> jangan import umap (hindari torch).
    - Jika aman (umap-learn), import umap.umap_ dan jalankan UMAP.
    """
    spec = importlib.util.find_spec("umap")
    if spec is None:
        raise ImportError(
            "Module 'umap' tidak ditemukan. Install 'umap-learn'. "
            "Perintah: pip uninstall -y umap && pip install -U umap-learn"
        )

    # Cari __init__.py tanpa mengimpor modul
    init_path = None

    # Pada banyak kasus, spec.origin menunjuk ke umap/__init__.py
    if getattr(spec, "origin", None):
        init_path = Path(spec.origin)
    else:
        # Jika origin None, coba ambil dari submodule_search_locations
        locs = getattr(spec, "submodule_search_locations", None)
        if locs:
            candidate = Path(list(locs)[0]) / "__init__.py"
            if candidate.exists():
                init_path = candidate

    content = ""
    if init_path and init_path.exists():
        try:
            content = init_path.read_text(encoding="utf-8", errors="ignore")
        except Exception:
            content = ""

    # Jika terdeteksi umap yang mengimport torch -> stop sebelum import umap
    if ("import torch" in content) or ("from torch" in content):
        raise ImportError(
            f"UMAP yang terdeteksi ({init_path}) mengimpor torch -> akan memicu error torch di environment kamu.\n"
            "Solusi environment yang benar: uninstall paket 'umap' yang salah, lalu install 'umap-learn':\n"
            "  pip uninstall -y umap\n"
            "  pip install -U umap-learn\n"
            "Restart kernel setelah itu.\n"
            "Saat ini kode akan fallback ke PCA agar tetap jalan."
        )

    # Aman: coba import umap-learn submodule
    umap_umap_ = importlib.import_module("umap.umap_")
    UMAP = getattr(umap_umap_, "UMAP", None)
    if UMAP is None:
        raise ImportError("Tidak menemukan umap.umap_.UMAP. Pastikan yang terinstal adalah 'umap-learn'.")

    return UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        random_state=random_state
    ).fit_transform(X)


def embed_2d(X, random_state=SEED):
    """
    Coba UMAP dulu tanpa memicu torch.
    Jika tidak bisa, fallback ke PCA.
    """
    try:
        return safe_umap_fit_transform(X, n_neighbors=15, min_dist=0.1, random_state=random_state)
    except Exception as e:
        print("\n[WARNING] UMAP tidak bisa digunakan tanpa memicu torch.")
        print("[WARNING] Alasan:", e)
        print("[WARNING] Fallback -> PCA untuk visualisasi.\n")
        return PCA(n_components=2, random_state=random_state).fit_transform(X)


# ======================
# DATA LOADING
# ======================
def load_perovskite_xrd_from_mp(
    api_key: str,
    max_per_class: int = MAX_PER_CLASS,
    two_theta_min: float = TWO_THETA_MIN,
    two_theta_max: float = TWO_THETA_MAX,
    n_points: int = N_POINTS,
    save_cif_dir: str | None = None,
):
    """
    Load perovskite from Materials Project, compute XRD, interpolate to fixed grid.
    Return: two_theta_grid, X (n_samples, n_points), y (string labels).
    """
    two_theta_grid = np.linspace(two_theta_min, two_theta_max, n_points)
    X, y = [], []
    class_counts = {c: 0 for c in TARGET_SYSTEMS}

    cif_dir = None
    if save_cif_dir is not None:
        cif_dir = Path(save_cif_dir)
        cif_dir.mkdir(parents=True, exist_ok=True)

    xrd_calc = XRDCalculator(wavelength="CuKa")

    with MPRester(api_key) as mpr:
        robo_docs = mpr.materials.robocrys.search(keywords=["perovskite"])
        perov_mpids = [str(doc.material_id) for doc in robo_docs][:10000]

        summaries = mpr.materials.summary.search(
            material_ids=perov_mpids,
            fields=["material_id", "structure", "symmetry"],
        )

        for doc in summaries:
            sym = getattr(doc, "symmetry", None)
            if sym is None or sym.crystal_system is None:
                continue

            cs_raw = sym.crystal_system
            if hasattr(cs_raw, "value"):
                cs = str(cs_raw.value).lower()
            else:
                cs_str = str(cs_raw)
                if "." in cs_str:
                    cs_str = cs_str.split(".")[-1]
                cs = cs_str.lower()

            if cs not in TARGET_SYSTEMS:
                continue
            if class_counts[cs] >= max_per_class:
                continue

            structure = doc.structure
            if structure is None:
                continue

            if cif_dir is not None:
                mpid = str(doc.material_id)
                cif_str = structure.to(fmt="cif")
                (cif_dir / f"{mpid}_{cs}.cif").write_text(cif_str)

            try:
                pattern = xrd_calc.get_pattern(
                    structure,
                    two_theta_range=(two_theta_min, two_theta_max),
                )
                intensities = np.interp(two_theta_grid, pattern.x, pattern.y)
                X.append(intensities)
                y.append(cs)
                class_counts[cs] += 1
            except Exception as e:
                print(f"Error calculating XRD for {doc.material_id}: {e}")
                continue

            if all(class_counts[c] >= max_per_class for c in TARGET_SYSTEMS):
                break

    X = np.array(X, dtype=np.float32)
    y = np.array(y)

    print("Jumlah sampel per kelas:", class_counts)
    if cif_dir is not None:
        print(f"File CIF disimpan di folder: {cif_dir.resolve()}")

    return two_theta_grid, X, y


# ======================
# RESAMPLING (TRAIN ONLY) - ADASYN
# ======================
def resample_train(X_train_raw, y_train, k_neighbors=5):
    k_neighbors = int(np.clip(k_neighbors, 3, 12))
    sampler = ADASYN(random_state=SEED, n_neighbors=k_neighbors)
    return sampler.fit_resample(X_train_raw, y_train)


# ======================
# CNN (HYPERPARAM)
# ======================
def build_cnn_1d_hp(input_shape, num_classes, hp: dict):
    """
    CNN hyperparams optimized by WOA.
    """
    l2_reg = float(hp.get("l2_reg", 0.0))
    bn = int(hp.get("batch_norm", 1))
    dr = float(hp.get("dropout", 0.0))

    reg = regularizers.l2(l2_reg) if l2_reg > 0 else None

    inputs = layers.Input(shape=input_shape)

    x = layers.Conv1D(int(hp["f1"]), int(hp["k1"]), padding="same", kernel_regularizer=reg)(inputs)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f2"]), int(hp["k2"]), padding="same", kernel_regularizer=reg)(x)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(int(hp["f3"]), int(hp["k3"]), padding="same", kernel_regularizer=reg)(x)
    if bn:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(int(hp["dense"]), activation="relu", kernel_regularizer=reg, name="feature_layer")(x)
    if dr > 0:
        x = layers.Dropout(dr)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)
    model = models.Model(inputs, outputs)

    loss_fn = make_sparse_ce_with_label_smoothing(num_classes, float(hp.get("label_smoothing", 0.0)))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=float(hp["lr"])),
        loss=loss_fn,
        metrics=["accuracy"],
    )
    return model


# ======================
# FEATURE / OVERLAP
# ======================
def get_feature_model(model):
    return Model(inputs=model.input, outputs=model.get_layer("feature_layer").output)


def safe_separability_metrics(features, y):
    sil = dbi = chi = None

    try:
        if len(np.unique(y)) >= 2 and features.shape[0] > len(np.unique(y)):
            sil = float(silhouette_score(features, y))
    except Exception:
        sil = None

    try:
        if len(np.unique(y)) >= 2:
            dbi = float(davies_bouldin_score(features, y))
    except Exception:
        dbi = None

    try:
        if len(np.unique(y)) >= 2:
            chi = float(calinski_harabasz_score(features, y))
    except Exception:
        chi = None

    return sil, dbi, chi


def overlap_penalty_from_features(features, y):
    """
    overlap penalty lebih ringan:
    - subsample max OVERLAP_MAX_POINTS untuk hitung silhouette/DBI
    """
    n = features.shape[0]
    if n > OVERLAP_MAX_POINTS:
        idx = np.random.choice(n, OVERLAP_MAX_POINTS, replace=False)
        features = features[idx]
        y = y[idx]

    sil, dbi, _ = safe_separability_metrics(features, y)

    if sil is None:
        sil_pen = 1.0
    else:
        sil_norm = (sil + 1.0) / 2.0
        sil_pen = 1.0 - sil_norm

    if dbi is None:
        dbi_pen = 1.0
    else:
        dbi_pen = float(np.log1p(dbi))

    return sil_pen + dbi_pen


# ======================
# WOA
# ======================
def decode_whale_position(pos, bounds):
    hp = {}
    for i, (name, (low, high, t)) in enumerate(bounds):
        x = float(pos[i])
        if t == "int":
            hp[name] = int(round(low + x * (high - low)))
        elif t == "float":
            hp[name] = float(low + x * (high - low))
        elif t == "float_log":
            log_low = math.log10(low)
            log_high = math.log10(high)
            hp[name] = 10 ** (log_low + x * (log_high - log_low))
        elif t == "bool":
            hp[name] = int(round(x))
        else:
            raise ValueError(f"Unknown type: {t}")

    # enforce odd kernels
    for k in ["k1", "k2", "k3"]:
        if hp[k] % 2 == 0:
            hp[k] += 1
        hp[k] = int(np.clip(hp[k], 3, 17))

    # clamp regs
    hp["dropout"] = float(np.clip(hp.get("dropout", 0.0), 0.0, 0.6))
    hp["label_smoothing"] = float(np.clip(hp.get("label_smoothing", 0.0), 0.0, 0.2))
    hp["l2_reg"] = float(np.clip(hp.get("l2_reg", 0.0), 0.0, 1e-2))
    hp["k_neighbors"] = int(np.clip(hp.get("k_neighbors", 5), 3, 12))
    return hp


def woa_optimize(X_train_raw, y_train, input_shape, num_classes):
    """
    WOA optimize CNN hp + k_neighbors ADASYN.
    bounds CNN diperkecil agar training kandidat cepat.
    """
    bounds = [
        ("lr", (5e-5, 2e-3, "float_log")),
        ("f1", (8, 48, "int")),
        ("k1", (3, 11, "int")),
        ("f2", (16, 96, "int")),
        ("k2", (3, 9, "int")),
        ("f3", (32, 128, "int")),
        ("k3", (3, 7, "int")),
        ("dense", (32, 160, "int")),
        ("dropout", (0.0, 0.5, "float")),
        ("l2_reg", (0.0, 8e-3, "float")),
        ("batch_norm", (0, 1, "bool")),
        ("label_smoothing", (0.0, 0.2, "float")),
        ("k_neighbors", (3, 12, "int")),
    ]

    dim = len(bounds)
    whales = np.random.rand(WOA_N_WHALES, dim)

    best_score = float("inf")
    best_hp = None
    best_pos = whales[0].copy()

    def fitness(pos):
        hp = decode_whale_position(pos, bounds)

        # 1) resample TRAIN ONLY
        X_res, y_res = resample_train(X_train_raw, y_train, k_neighbors=hp["k_neighbors"])

        # 2) split resampled train -> train_sub/val
        X_tr, X_val, y_tr, y_val = train_test_split(
            X_res, y_res, test_size=WOA_VAL_SIZE, random_state=SEED, stratify=y_res
        )

        # 3) add channel
        X_tr_c = X_tr[..., np.newaxis]
        X_val_c = X_val[..., np.newaxis]

        # 4) train quick model
        model = build_cnn_1d_hp(input_shape, num_classes, hp)

        es = tf.keras.callbacks.EarlyStopping(
            monitor="val_loss", patience=1, restore_best_weights=True
        )

        hist = model.fit(
            X_tr_c, y_tr,
            epochs=WOA_EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_val_c, y_val),
            verbose=0,
            callbacks=[es]
        )

        val_loss = float(np.min(hist.history["val_loss"]))
        y_val_pred = np.argmax(model.predict(X_val_c, verbose=0), axis=1)
        val_macro_f1 = float(f1_score(y_val, y_val_pred, average="macro", zero_division=0))

        # overlap penalty from feature space (val only)
        feat_model = get_feature_model(model)
        val_features = feat_model.predict(X_val_c, verbose=0, batch_size=128)
        overlap_pen = float(overlap_penalty_from_features(val_features, y_val))

        score = val_loss + LAMBDA_F1 * (1.0 - val_macro_f1) + LAMBDA_OVERLAP * overlap_pen

        tf.keras.backend.clear_session()
        return score, hp

    # initial eval
    for i in range(WOA_N_WHALES):
        sc, hp = fitness(whales[i])
        if sc < best_score:
            best_score, best_hp, best_pos = sc, hp, whales[i].copy()

    # iterate
    for t in range(WOA_N_ITERS):
        a = 2 - t * (2 / WOA_N_ITERS)

        for i in range(WOA_N_WHALES):
            r1 = np.random.rand(dim)
            r2 = np.random.rand(dim)
            A = 2 * a * r1 - a
            C = 2 * r2
            p = np.random.rand()

            if p < 0.5:
                if np.linalg.norm(A) < 1:
                    D = np.abs(C * best_pos - whales[i])
                    new_pos = best_pos - A * D
                else:
                    rand_idx = np.random.randint(WOA_N_WHALES)
                    X_rand = whales[rand_idx]
                    D = np.abs(C * X_rand - whales[i])
                    new_pos = X_rand - A * D
            else:
                D = np.abs(best_pos - whales[i])
                b = 1
                l = np.random.uniform(-1, 1, dim)
                new_pos = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_pos

            whales[i] = np.clip(new_pos, 0, 1)

        for i in range(WOA_N_WHALES):
            sc, hp = fitness(whales[i])
            if sc < best_score:
                best_score, best_hp, best_pos = sc, hp, whales[i].copy()

        print(f"[WOA-{STRATEGY_NAME}] iter {t+1}/{WOA_N_ITERS} best_score={best_score:.5f}")
        print(f"   best_hp={best_hp}")

    return best_hp, best_score


# ======================
# PLOTS
# ======================
def plot_confusion_matrix(y_true, y_pred, class_names, title, save_path=None):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title, fontweight="bold")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))


def plot_umap_raw_vs_features_single(X_raw, features, y_encoded, class_names, title_prefix, save_path):
    """
    UMAP tetap ada:
    - Coba UMAP aman (tidak memicu torch)
    - Kalau gagal -> PCA
    """
    n_samples = min(UMAP_N_SAMPLES, len(X_raw))
    if len(X_raw) > n_samples:
        sel = np.random.choice(len(X_raw), n_samples, replace=False)
        X_raw_s = X_raw[sel]
        feats_s = features[sel]
        y_s = y_encoded[sel]
    else:
        X_raw_s, feats_s, y_s = X_raw, features, y_encoded

    emb_raw = embed_2d(X_raw_s, random_state=SEED)
    emb_feat = embed_2d(feats_s, random_state=SEED)

    colors = ["red", "blue", "green", "orange", "purple", "brown"]
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    for i, lab in enumerate(np.unique(y_s)):
        m = y_s == lab
        axes[0].scatter(emb_raw[m, 0], emb_raw[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)
        axes[1].scatter(emb_feat[m, 0], emb_feat[m, 1], s=20, alpha=0.65,
                        c=colors[i % len(colors)], label=class_names[lab],
                        edgecolors="black", linewidths=0.15)

    axes[0].set_title(f"{title_prefix} - Raw (2048D→2D)", fontweight="bold")
    axes[1].set_title(f"{title_prefix} - Features (→2D)", fontweight="bold")
    axes[0].grid(True, alpha=0.3)
    axes[1].grid(True, alpha=0.3)
    axes[0].legend(fontsize=8)
    axes[1].legend(fontsize=8)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()


# ======================
# MAIN
# ======================
if __name__ == "__main__":
    set_seed(SEED)

    print("=" * 70)
    print(f"WOA + CNN PIPELINE (ADASYN) - Revised (No Torch Import) | WOA iters={WOA_N_ITERS}")
    print("=" * 70)

    two_theta, X_raw, y_str = load_perovskite_xrd_from_mp(API_KEY, save_cif_dir=str(CIF_SAVE_DIR))

    le = LabelEncoder()
    y_encoded = le.fit_transform(y_str)
    class_names = list(le.classes_)

    # split train/test ONCE
    X_train_raw, X_test_raw, y_train, y_test = train_test_split(
        X_raw, y_encoded, test_size=0.2, random_state=SEED, stratify=y_encoded
    )

    print(f"Train raw: {len(X_train_raw)} | Test raw: {len(X_test_raw)}")
    print("Train dist(raw):", Counter(y_train))
    print("Test dist:", Counter(y_test))

    input_shape = (X_train_raw.shape[1], 1)
    num_classes = len(class_names)

    # WOA optimize
    best_hp, best_score = woa_optimize(X_train_raw, y_train, input_shape, num_classes)

    print("\n=== BEST FOUND (WOA) ===")
    print("Best score:", best_score)
    print("Best hp:", best_hp)

    # final resample train with best k
    X_train_res, y_train_res = resample_train(
        X_train_raw, y_train, k_neighbors=best_hp["k_neighbors"]
    )

    print("\nAfter ADASYN (final):")
    print("Train size:", len(X_train_res))
    print("Train dist:", Counter(y_train_res))

    X_train = X_train_res[..., np.newaxis]
    X_test = X_test_raw[..., np.newaxis]

    # final training
    model = build_cnn_1d_hp(input_shape, num_classes, best_hp)
    es = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)

    history = model.fit(
        X_train, y_train_res,
        epochs=EPOCHS_FINAL,
        batch_size=BATCH_SIZE,
        validation_split=0.2,
        verbose=1,
        callbacks=[es]
    )

    # test evaluation
    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    print("\n=== TEST RESULTS ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")

    # extract features for all data (untuk UMAP/PCA plot)
    feat_model = get_feature_model(model)
    features_all = feat_model.predict(X_raw[..., np.newaxis], verbose=0, batch_size=128)

    # save best hp
    with open(RESULT_DIR / "best_hp.json", "w", encoding="utf-8") as f:
        json.dump(best_hp, f, indent=2)

    # confusion matrix
    plot_confusion_matrix(
        y_test, y_pred, class_names,
        title=f"Confusion Matrix - {STRATEGY_NAME} (WOA+CNN)",
        save_path=str(RESULT_DIR / "cm.png")
    )

    # UMAP (atau PCA fallback)
    plot_umap_raw_vs_features_single(
        X_raw=X_raw,
        features=features_all,
        y_encoded=y_encoded,
        class_names=class_names,
        title_prefix=STRATEGY_NAME,
        save_path=str(RESULT_DIR / "umap_raw_vs_features.png")
    )

    tf.keras.backend.clear_session()

    print("\nSaved outputs in:", RESULT_DIR.resolve())
    print("DONE.")
```
