```python
# ============================================================================
# DATA LOADING FROM MATERIALS PROJECT
# ============================================================================

def load_perovskite_data_from_mp(api_key: str, max_per_class: int = MAX_PER_CLASS):
    """Load perovskite data from Materials Project"""

    data_records = []
    class_counts = {c: 0 for c in TARGET_SYSTEMS}

    with MPRester(api_key) as mpr:
        # Search for perovskite materials
        robo_docs = mpr.materials.robocrys.search(keywords=["perovskite"])
        perov_mpids = [str(doc.material_id) for doc in robo_docs][:50000]

        print(f"Found {len(perov_mpids)} perovskite materials")

        # Get material summaries
        summaries = mpr.materials.summary.search(
            material_ids=perov_mpids,
            fields=["material_id", "structure", "symmetry", "formula_pretty",
                   "density", "volume", "nsites"],
        )

        for doc in summaries:
            sym = getattr(doc, "symmetry", None)
            if sym is None or sym.crystal_system is None:
                continue

            cs_raw = sym.crystal_system
            if hasattr(cs_raw, "value"):
                cs = str(cs_raw.value).lower()
            else:
                cs_str = str(cs_raw)
                if "." in cs_str:
                    cs_str = cs_str.split(".")[-1]
                cs = cs_str.lower()

            if cs not in TARGET_SYSTEMS:
                continue
            if class_counts[cs] >= max_per_class:
                continue

            structure = doc.structure
            if structure is None:
                continue

            # Collect data
            record = {
                'material_id': str(doc.material_id),
                'formula': str(doc.formula_pretty) if hasattr(doc, 'formula_pretty') else None,
                'crystal_system': cs,
                'density': float(doc.density) if hasattr(doc, 'density') else None,
                'volume': float(doc.volume) if hasattr(doc, 'volume') else None,
                'nsites': int(doc.nsites) if hasattr(doc, 'nsites') else None,
                'structure': structure
            }

            data_records.append(record)
            class_counts[cs] += 1

            if all(class_counts[c] >= max_per_class for c in TARGET_SYSTEMS):
                break

    print(f"\nTotal materials collected: {len(data_records)}")
    print(f"Class distribution: {class_counts}")

    return pd.DataFrame(data_records)

# Load data
print("\n" + "="*70)
print("LOADING DATA FROM MATERIALS PROJECT")
print("="*70)
df_materials = load_perovskite_data_from_mp(API_KEY)
print(f"\nDataset shape: {df_materials.shape}")
```

```python
# ============================================================================
# XRD PATTERN GENERATION
# ============================================================================

def generate_xrd_patterns(df_materials, two_theta_min=TWO_THETA_MIN,
                         two_theta_max=TWO_THETA_MAX, n_points=N_POINTS):
    """Generate XRD patterns from structures"""

    two_theta_grid = np.linspace(two_theta_min, two_theta_max, n_points)
    X_patterns = []
    y_systems = []
    material_ids = []  # Store material IDs
    formulas = []  # Store formulas
    valid_indices = []
    failed_count = 0

    xrd_calc = XRDCalculator(wavelength="CuKa")

    print("="*70)
    print("GENERATING XRD PATTERNS")
    print("="*70)
    print(f"2θ range: {two_theta_min}° - {two_theta_max}°")
    print(f"Number of points: {n_points}")
    print(f"Wavelength: Cu Kα")
    print()

    total = len(df_materials)

    for idx, row in df_materials.iterrows():
        try:
            structure = row['structure']
            pattern = xrd_calc.get_pattern(
                structure,
                two_theta_range=(two_theta_min, two_theta_max),
            )
            intensities = np.interp(two_theta_grid, pattern.x, pattern.y)

            X_patterns.append(intensities)
            y_systems.append(row['crystal_system'])
            material_ids.append(row['material_id'])
            formulas.append(row['formula'])
            valid_indices.append(idx)

        except Exception as e:
            failed_count += 1
            if failed_count <= 3:  # Only show first 3 errors
                print(f"✗ Error for {row['material_id']}: {str(e)[:50]}...")
            continue

        if (len(X_patterns) % 500) == 0:
            print(f"✓ Processed {len(X_patterns)}/{total} patterns... ({len(X_patterns)/total*100:.1f}%)")

    X_patterns = np.array(X_patterns, dtype=np.float32)
    y_systems = np.array(y_systems)
    material_ids = np.array(material_ids)
    formulas = np.array(formulas)

    print()
    print("="*70)
    print("XRD GENERATION SUMMARY")
    print("="*70)
    print(f"✓ Successfully generated: {len(X_patterns)} patterns")
    print(f"✗ Failed: {failed_count} patterns")
    print(f"Success rate: {len(X_patterns)/(len(X_patterns)+failed_count)*100:.2f}%")
    print(f"Pattern shape: {X_patterns.shape}")
    print(f"Crystal system distribution:\n{Counter(y_systems)}")

    return two_theta_grid, X_patterns, y_systems, material_ids, formulas, valid_indices

# Generate XRD patterns
two_theta_grid, X_raw, y_str, material_ids, formulas, valid_idx = generate_xrd_patterns(df_materials)

# Update dataframe to keep only valid materials
df_materials = df_materials.iloc[valid_idx].reset_index(drop=True)

print(f"\n✓ Final dataset size: {len(X_raw)} samples")
```

```python
# ============================================================================
# SAFE-LEVEL SMOTE - CLASS DEFINITION (AGGRESSIVE MODE)
# ============================================================================
# Referensi:
#   Bunkhumpornpat, C., Sinapiromsaran, K., & Lursinsap, C. (2009).
#   Safe-Level-SMOTE: Safe-Level-Synthetic Minority Over-Sampling TEchnique
#   for Handling the Class Imbalanced Problem. PAKDD 2009.
#
# Pengaturan agresif untuk menghindari overlap:
#
#   Parameter kunci:
#   ┌─────────────────┬─────────────────────────────────────────────────────┐
#   │ k_neighbors     │ Jumlah tetangga untuk hitung safe level.            │
#   │                 │ Lebih kecil → deteksi overlap lebih lokal & sensitif│
#   ├─────────────────┼─────────────────────────────────────────────────────┤
#   │ min_safe_level  │ Ambang minimum safe level agar sampel boleh jadi    │
#   │                 │ seed interpolasi.                                   │
#   │                 │ • Default paper : 1  (butuh ≥1/k tetangga sekelas)  │
#   │                 │ • Rekomendasi   : 3  (butuh ≥3/k = ≥60% sekelas)   │
#   │                 │ → Berlaku untuk semua kelas: mayoritas-minoritas    │
#   │                 │   DAN minoritas-minoritas                           │
#   ├─────────────────┼─────────────────────────────────────────────────────┤
#   │ gap_max_ratio   │ Batas atas gap interpolasi.                         │
#   │                 │ • Default paper: 1.0 → synthetic bisa sejauh xnn   │
#   │                 │ • Agresif      : 0.5 → synthetic max di titik       │
#   │                 │   TENGAH xi→xnn, tidak melewati batas kelas        │
#   └─────────────────┴─────────────────────────────────────────────────────┘
#
#   Kombinasi rekomendasi agresif:
#     k_neighbors=5, min_safe_level=3, gap_max_ratio=0.5
#
#   Artinya:
#     • Seed hanya dari sampel dengan ≥3/5 tetangga sekelas
#       → menjamin seed berada jauh dari batas kelas manapun
#     • Synthetic hanya dibuat di separuh pertama segmen xi→xnn
#       → tidak melewati titik tengah → tidak masuk wilayah kelas lain
# ============================================================================

class SafeLevelSMOTE:
    """
    Safe-Level SMOTE dengan parameter agresif untuk menghindari overlap.

    Parameters
    ----------
    sampling_strategy : 'auto' atau dict
        'auto' → oversample semua kelas minoritas ke jumlah kelas mayoritas.
    k_neighbors : int, default=5
        Jumlah tetangga untuk menghitung safe level.
        Safe level dihitung dari SELURUH training set sehingga
        overlap antar kelas minoritas ikut dipertimbangkan.
    min_safe_level : int, default=3
        Ambang minimum safe level agar sampel boleh jadi seed.
        Nilai 3 dengan k=5 → butuh ≥60% tetangga sekelas.
        Semakin tinggi → semakin agresif. Rentang valid: 1 s/d k_neighbors.
    gap_max_ratio : float, default=0.5
        Batas atas gap interpolasi [0, gap_max_ratio * effective_max].
        Nilai 0.5 → synthetic maksimal di titik TENGAH xi→xnn.
        Rentang valid: 0.0 s/d 1.0.
    random_state : int atau None
    """

    def __init__(self, sampling_strategy='auto', k_neighbors=5,
                 min_safe_level=4, gap_max_ratio=0.5, random_state=None):
        self.sampling_strategy = sampling_strategy
        self.k_neighbors       = k_neighbors
        self.min_safe_level    = min_safe_level
        self.gap_max_ratio     = gap_max_ratio
        self.random_state      = random_state

    def fit_resample(self, X, y):
        rng = np.random.RandomState(self.random_state)
        X   = np.array(X, dtype=np.float32)
        y   = np.array(y)

        class_counts   = Counter(y)
        majority_count = max(class_counts.values())

        if self.sampling_strategy == 'auto':
            target_counts = {cls: majority_count
                             for cls, cnt in class_counts.items()
                             if cnt < majority_count}
        else:
            target_counts = self.sampling_strategy

        # ── Safe level dihitung dari SELURUH data (semua kelas) ───────────
        # Jika xi dikelilingi kelas lain (mayoritas ATAU minoritas berbeda),
        # safe level-nya rendah dan tidak akan jadi seed.
        k_fit  = min(self.k_neighbors + 1, len(X))
        nn_all = NearestNeighbors(n_neighbors=k_fit, algorithm='auto', n_jobs=-1)
        nn_all.fit(X)
        all_neighbors = nn_all.kneighbors(X, return_distance=False)

        X_syn_list = []
        y_syn_list = []

        for class_label, target_count in target_counts.items():
            n_generate = target_count - class_counts[class_label]
            if n_generate <= 0:
                continue

            minority_idx = np.where(y == class_label)[0]

            safe_levels = np.array([
                np.sum(y[all_neighbors[idx, 1:self.k_neighbors + 1]] == class_label)
                for idx in minority_idx
            ])

            # ── Filter agresif: sl >= min_safe_level ──────────────────────
            # Menolak sampel yang dekat dengan batas kelas MANAPUN
            safe_mask = safe_levels >= self.min_safe_level
            safe_idx  = minority_idx[safe_mask]
            safe_sl   = safe_levels[safe_mask]

            n_qualified   = len(safe_idx)
            n_total_min   = len(minority_idx)
            pct_qualified = (n_qualified / n_total_min * 100) if n_total_min > 0 else 0
            print(f"  [{class_label}] Safe seeds: {n_qualified}/{n_total_min} "                  f"({pct_qualified:.1f}%) dengan sl >= {self.min_safe_level}")

            if n_qualified == 0:
                # Fallback: turunkan ke sl >= 1 agar kelas tidak terlewat
                print(f"  ⚠ Fallback ke sl>=1 untuk kelas '{class_label}'")
                safe_mask = safe_levels >= 1
                safe_idx  = minority_idx[safe_mask]
                safe_sl   = safe_levels[safe_mask]
                if len(safe_idx) == 0:
                    print(f"  ⚠ Semua sampel kelas '{class_label}' noise, dilewati.")
                    continue

            k_min    = min(self.k_neighbors, len(minority_idx) - 1)
            k_min    = max(k_min, 1)
            nn_minor = NearestNeighbors(n_neighbors=k_min + 1, algorithm='auto', n_jobs=-1)
            nn_minor.fit(X[minority_idx])
            minor_neighbors = nn_minor.kneighbors(X[safe_idx], return_distance=False)

            sl_lookup = dict(zip(minority_idx.tolist(), safe_levels.tolist()))

            X_syn = np.empty((n_generate, X.shape[1]), dtype=np.float32)

            for i in range(n_generate):
                pick   = rng.randint(0, len(safe_idx))
                xi_idx = safe_idx[pick]
                sl_xi  = int(safe_sl[pick])

                nn_row  = minor_neighbors[pick, 1:]
                nn_orig = minority_idx[nn_row]
                nn_orig = nn_orig[nn_orig != xi_idx]
                if len(nn_orig) == 0:
                    X_syn[i] = X[xi_idx]
                    continue

                xnn_idx = rng.choice(nn_orig)
                sl_xnn  = int(sl_lookup.get(int(xnn_idx), 0))

                if sl_xnn == 0:
                    X_syn[i] = X[xi_idx]
                else:
                    sl_ratio = sl_xi / sl_xnn

                    # Hitung batas atas gap berdasarkan sl-ratio
                    if sl_ratio > 1.0:
                        raw_max = 1.0 / sl_ratio   # xi lebih aman → dekat xi
                    else:
                        raw_max = 1.0              # xnn sama/lebih aman

                    # ── gap_max_ratio: batasi lebih lanjut agar synthetic
                    #    tidak melewati titik tengah (gap_max_ratio=0.5) ──
                    effective_max = min(raw_max, self.gap_max_ratio)
                    gap = rng.uniform(0.0, effective_max)
                    X_syn[i] = X[xi_idx] + gap * (X[xnn_idx] - X[xi_idx])

            X_syn_list.append(X_syn)
            y_syn_list.extend([class_label] * n_generate)

        if X_syn_list:
            X_resampled = np.vstack([X] + X_syn_list)
            y_resampled = np.append(y, y_syn_list)
        else:
            X_resampled = X.copy()
            y_resampled = y.copy()

        return X_resampled, y_resampled


print("✓ SafeLevelSMOTE (Aggressive Mode) berhasil didefinisikan.")
print()
print("  Tabel konfigurasi agresif:")
print("  ┌───────────────────┬────────┬──────────────────────────────────────┐")
print("  │ Parameter         │ Nilai  │ Efek                                 │")
print("  ├───────────────────┼────────┼──────────────────────────────────────┤")
print("  │ k_neighbors       │   5    │ 5 tetangga dari seluruh dataset      │")
print("  │ min_safe_level    │   3    │ seed hanya jika ≥3/5 tetangga sekelas│")
print("  │ gap_max_ratio     │  0.5   │ synthetic max di titik tengah xi→xnn │")
print("  └───────────────────┴────────┴──────────────────────────────────────┘")
print("  → Overlap minoritas-mayoritas   : dihindari via min_safe_level")
print("  → Overlap minoritas-minoritas   : dihindari via min_safe_level")
print("  → Synthetic tidak melintasi batas: dijamin via gap_max_ratio=0.5")
```

```python
# ============================================================================
# APPLYING SAFE-LEVEL SMOTE (AGGRESSIVE - AUTO BALANCE ALL CLASSES)
# ============================================================================

print("\n" + "="*70)
print("APPLYING SAFE-LEVEL SMOTE (AGGRESSIVE MODE)")
print("="*70)
print("Konfigurasi agresif untuk menghindari overlap:")
print("  k_neighbors    = 5    → 5 tetangga dari seluruh dataset")
print("  min_safe_level = 3    → seed hanya jika ≥3/5 (60%) tetangga sekelas")
print("                          berlaku terhadap kelas MANAPUN:")
print("                          ✓ minoritas-mayoritas")
print("                          ✓ minoritas-minoritas")
print("  gap_max_ratio  = 0.5  → synthetic max di titik TENGAH xi→xnn")
print("                          tidak melintasi batas kelas")
print()

print("BEFORE Safe-Level SMOTE:")
print("-" * 50)
print(f"Training samples: {len(X_train_raw)}")
print("\nClass distribution:")
for i, name in enumerate(class_names):
    print(f"  {name.capitalize():<15}: {np.sum(y_train == i):>5} samples")
print()

sampling_strategy_dict = {
    0: 3414,   # cubic        → isi jumlah aslinya saja (mayoritas, tidak diubah)
    1: 760,   # monoclinic   → target jumlah akhir
    2: 780,   # orthorhombic → target jumlah akhir
    3: 740,   # tetragonal   → target jumlah akhir
}

sl_smote = SafeLevelSMOTE(
    sampling_strategy=sampling_strategy_dict,  # oversample semua minoritas ke jumlah mayoritas
    k_neighbors=2,             # tetangga untuk safe level (seluruh dataset)
    min_safe_level=2,          # ≥3/5 tetangga harus sekelas → filter agresif
    gap_max_ratio=0.5,         # synthetic max di titik tengah → tidak overlap
    random_state=SEED
)

print("Fitting Safe-Level SMOTE...")
print("Safe seed stats per kelas:")
X_train_smote, y_train_smote = sl_smote.fit_resample(X_train_raw, y_train)

print("\n" + "="*70)
print("AFTER Safe-Level SMOTE:")
print("-" * 50)
print(f"Training samples: {len(X_train_smote)} (changed by {len(X_train_smote) - len(X_train_raw):+d})")
print("\nClass distribution:")
for i, name in enumerate(class_names):
    count          = np.sum(y_train_smote == i)
    original_count = np.sum(y_train == i)
    diff           = count - original_count
    diff_pct       = (diff / original_count * 100) if original_count > 0 else 0
    sign           = '+' if diff >= 0 else ''
    print(f"  {name.capitalize():<15}: {count:>5} samples (Original: {original_count}, {sign}{diff} = {sign}{diff_pct:.1f}%)")

print()
print("  min_safe_level=3 → sampel di batas kelas manapun dilewati sebagai seed")
print("  gap_max_ratio=0.5 → synthetic tidak melewati titik tengah xi→xnn")

print("\n" + "="*70)
print("VALIDATION SET (UNCHANGED):")
print("-" * 50)
for i, name in enumerate(class_names):
    print(f"  {name.capitalize():<15}: {np.sum(y_val == i):>5} samples")

print("\n" + "="*70)
print("TEST SET (UNCHANGED):")
print("-" * 50)
for i, name in enumerate(class_names):
    print(f"  {name.capitalize():<15}: {np.sum(y_test == i):>5} samples")

print("\n✓ Safe-Level SMOTE (Aggressive) applied successfully!")
```

```python
# ============================================================================
# MODEL ARCHITECTURE DEFINITIONS
# ============================================================================

def build_cnn_fixed(input_shape, num_classes: int):
    """
    Build CNN with fixed architecture
    - 3 Conv1D layers: filters=80, kernel_size=[100, 80, 20], stride=5
    - Average Pooling after each conv layer
    - 2 Fully Connected layers with 30% Dropout
    """
    inputs = layers.Input(shape=input_shape)

    # Convolutional Layer 1: 80 filters, kernel_size=100, stride=5
    x = layers.Conv1D(filters=80, kernel_size=80, strides=5,
                      activation="relu", padding="same",
                      name="conv1")(inputs)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool1")(x)

    # Convolutional Layer 2: 80 filters, kernel_size=80, stride=5
    x = layers.Conv1D(filters=80, kernel_size=40, strides=5,
                      activation="relu", padding="same",
                      name="conv2")(x)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool2")(x)

    # Convolutional Layer 3: 40 filters, kernel_size=20, stride=5
    x = layers.Conv1D(filters=80, kernel_size=20, strides=5,
                      activation="relu", padding="same",
                      name="conv3")(x)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool3")(x)


    # Flatten
    x = layers.Flatten()(x)

    # Fully Connected Layer 1 with Dropout 30%
    x = layers.Dense(128, activation="relu", name="fc1")(x)
    x = layers.Dropout(0.3, name="dropout1")(x)

    # Fully Connected Layer 2 (Feature Layer) with Dropout 30%
    x = layers.Dense(64, activation="relu", name="feature_layer")(x)
    x = layers.Dropout(0.3, name="dropout2")(x)

    # Output Layer
    outputs = layers.Dense(num_classes, activation="softmax", name="output")(x)

    model = models.Model(inputs, outputs, name="CNN_Fixed")

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )

    return model


def build_cnn_woa(params, input_shape, num_classes):
    """
    Build CNN with WOA-optimized hyperparameters
    Architecture:
    - 4 Conv1D layers: all with kernel_size optimized, stride=5
    - Average Pooling after each conv layer
    - 2 Fully Connected layers with 30% Dropout

    WOA optimizes:
    - Number of filters for all 3 conv layers (same value)
    - Fully connected layer 1 units
    - Fully connected layer 2 units (feature layer)
    """
    num_filters = int(params[0])      # Filters for all 3 conv layers
    fc1_units = int(params[1])        # FC layer 1 units
    fc2_units = int(params[2])        # FC layer 2 units (feature layer)

    inputs = layers.Input(shape=input_shape)

    # Convolutional Layer 1: kernel_size=100, stride=5
    x = layers.Conv1D(filters=num_filters, kernel_size=100, strides=5,
                      activation="relu", padding="same",
                      name="conv1")(inputs)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool1")(x)

    # Convolutional Layer 2: kernel_size=80, stride=5
    x = layers.Conv1D(filters=num_filters, kernel_size=70, strides=5,
                      activation="relu", padding="same",
                      name="conv2")(x)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool2")(x)

    # Convolutional Layer 3: kernel_size=20, stride=5
    x = layers.Conv1D(filters=num_filters, kernel_size=40, strides=5,
                      activation="relu", padding="same",
                      name="conv3")(x)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool3")(x)

    # Convolutional Layer 4: kernel_size=20, stride=5
    x = layers.Conv1D(filters=num_filters, kernel_size=10, strides=5,
                      activation="relu", padding="same",
                      name="conv4")(x)
    x = layers.AveragePooling1D(pool_size=2, name="avgpool4")(x)

    # Flatten
    x = layers.Flatten()(x)

    # Fully Connected Layer 1 with Dropout 30%
    x = layers.Dense(fc1_units, activation="relu", name="fc1")(x)
    x = layers.Dropout(0.3, name="dropout1")(x)

    # Fully Connected Layer 2 (Feature Layer) with Dropout 30%
    x = layers.Dense(fc2_units, activation="relu", name="feature_layer")(x)
    x = layers.Dropout(0.3, name="dropout2")(x)

    # Output Layer
    outputs = layers.Dense(num_classes, activation="softmax", name="output")(x)

    model = models.Model(inputs, outputs, name="CNN_WOA")

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )

    return model
```

```python
# ============================================================================
# t-SNE VISUALIZATION FOR ALL MODELS
# ============================================================================

print("\n" + "="*70)
print("GENERATING t-SNE VISUALIZATIONS")
print("="*70)

tsne = TSNE(n_components=2, random_state=SEED, perplexity=30, n_iter=1000, verbose=0)

print("Applying t-SNE to CNN (No Safe-Level SMOTE) features...")
X_tsne_cnn_no_smote = tsne.fit_transform(features_cnn_no_smote)

print("Applying t-SNE to CNN (With Safe-Level SMOTE) features...")
X_tsne_cnn_with_smote = tsne.fit_transform(features_cnn_with_smote)

print("✓ t-SNE transformation complete for all models!")
```
